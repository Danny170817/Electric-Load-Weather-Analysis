```python
#import libariries
import pandas as pd
import numpy as np
```

# Data Understanding & Preparation

### Load the Dataset

```python
# Load the CSV files
load_data = pd.read_csv("load_data.csv")
weather_data = pd.read_csv("weather_data.csv")
```

### Step 1.1: Inspecting First Rows

```python
# Display the first few rows of each dataset
print("Load Data:")
display(load_data.head())

print("Weather Data:")
display(weather_data.head())

# Check dataset info
print("\nLoad Data Info:")
load_data.info()

print("\n Weather Data Info:")
weather_data.info()

# Check missing values
print("\n‚ùì Missing Values in Load Data:")
print(load_data.isnull().sum())

print("\n‚ùì Missing Values in Weather Data:")
print(weather_data.isnull().sum())
```

```python
#Generate descriptive statistics.

print(load_data.describe())

print(weather_data.describe())
```

### Reshape the Data

```python
# Convert from wide to long format
load_long = load_data.melt(id_vars=["zone_id", "year", "month", "day"], 
                      var_name="hour", value_name="load_MW")

weather_long = weather_data.melt(id_vars=["station_id", "year", "month", "day"], 
                            var_name="hour", value_name="temperature")
```

```python
print("Unique hour values (load):", load_long["hour"].unique())
```

###  Fixing hours and datetime

```python
# Convert hour column to string first
load_long["hour"] = load_long["hour"].astype(str).str.extract("h(\d+)").astype(float)
weather_long["hour"] = weather_long["hour"].astype(str).str.extract("h(\d+)").astype(float)


# Create full date and time column
load_long["datetime"] = pd.to_datetime(load_long[["year", "month", "day"]]) + pd.to_timedelta(load_long["hour"] - 1, unit="h")
weather_long["datetime"] = pd.to_datetime(weather_long[["year", "month", "day"]]) + pd.to_timedelta(weather_long["hour"] - 1, unit="h")
```

```python
load_long
```

### merging of data

```python
# Combine the two datasets on datetime
df = pd.merge(load_long, weather_long[["datetime", "temperature"]], on="datetime")

# Remove rows with missing values
df = df.dropna()

# Look at the result
df.head()
```

```python
# Add extra time-related columns
df["weekday"] = df["datetime"].dt.day_name()
df["hour"] = df["datetime"].dt.hour
df["month"] = df["datetime"].dt.month
```

```python
#displaying data
print("Data:", df)
```

## 2. Exploratory Data Analysis


```python
import matplotlib.pyplot as plt

# Plot load trend over time
plt.figure(figsize=(15, 5))
plt.plot(df["datetime"], df["load_MW"], color="blue", alpha=0.7)
plt.title("Hourly Electric Load Over Time")
plt.xlabel("Date")
plt.ylabel("Load (MW)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

Seasonal Patterns
Regular yearly cycles are clearly visible.

Higher loads during summer and winter months, possibly due to:

Air conditioning in summer

Heating in winter
Lower demand in milder months

There's a slight upward trend in base load over the years.

Indicates growing energy consumption 

```python
# Average load by hour
hourly_avg = df.groupby("hour")["load_MW"].mean()

plt.figure(figsize=(10, 4))
plt.plot(hourly_avg.index, hourly_avg.values, marker='o')
plt.title("Average Load by Hour of Day")
plt.xlabel("Hour")
plt.ylabel("Average Load (MW)")
plt.grid(True)
plt.xticks(range(0, 24))
plt.tight_layout()
plt.show()
```

Pattern Observed:
Low demand during the early morning hours (midnight to 5 AM)

Sharp rise starts around 6 AM, likely as people wake up and businesses begin operating

Two peaks observed:

Morning peak: Around 7‚Äì9 AM

Evening peak: Highest between 6‚Äì8 PM (around 18:00‚Äì20:00)

Drop after 9 PM as daily activity slows down

```python
weekday_avg = df.groupby("weekday")["load_MW"].mean()

plt.figure(figsize=(10, 4))
plt.bar(weekday_avg.index, weekday_avg.values, color='orange')
plt.title("Average Load by Day of Week")
plt.ylabel("Average Load (MW)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Load values are very similar across all days of the week.

No significant drop on weekends (Saturday, Sunday), which is often seen in residential-focused grids.

```python
# Scatter plot to see relationship
plt.figure(figsize=(8, 5))
plt.scatter(df["temperature"], df["load_MW"], alpha=0.3)
plt.title("Load vs Temperature")
plt.xlabel("Temperature (¬∞F or ¬∞C depending on data)")
plt.ylabel("Load (MW)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

Higher load at low temperatures (left side of U): Increased heating demand 

Higher load at high temperatures (right side of U):Increased cooling demand (air conditioners, refrigeration)

```python
# Quick correlation
correlation = df[["load_MW", "temperature"]].corr()
print("Correlation between Load and Temperature:\n", correlation)
```

Correlation coefficient: -0.094263


## Linear Regression 

```python
from sklearn.linear_model import LinearRegression
X = df[["temperature"]]  
y = df["load_MW"]  
```

```python
model = LinearRegression()
model.fit(X, y)  # train the model
```

```python
# Predict load using the same temperature data
y_pred = model.predict(X)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.scatter(df["temperature"], df["load_MW"], alpha=0.3, label="Actual")
plt.plot(df["temperature"], y_pred, color="red", label="Predicted Line")
plt.title("Linear Regression: Load vs Temperature")
plt.xlabel("Temperature")
plt.ylabel("Load (MW)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

I created a simple linear regression model using temperature to predict electric load. The model captures basic trends, but since the relationship is not linear, a more advanced model
