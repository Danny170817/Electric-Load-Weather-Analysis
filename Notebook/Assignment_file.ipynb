```python
load_data=[]
hour_cols=[col for col in load if col.startswith('h')]

for index,row in load.iterrows():
    for hour_col in hour_cols:
        load_data.append({
            'zone_id':row['zone_id'],
            'year':row['year'],
            'month':row['month'],
            'day':row['day'],
            'hour':hour_col,
            'load_MW':row[hour_col],
        })


load_data_loop=pd.DataFrame(load_data)
load_data_loop
```

```python
weather_data=[]
weather_hour_cols=[col for col in load if col.startswith('h')]

for index,row in weather.iterrows():
    for hour_col in hour_cols:
        weather_data.append({
            'station_id':row['station_id'],
            'year':row['year'],
            'month':row['month'],
            'day':row['day'],
            'hour':hour_col,
            'temperature':row[hour_col],
        })


weather_data_loop=pd.DataFrame(weather_data)
weather_data_loop
print(weather_data_loop.shape)
```

```python
print("Unique hour values (load):", load_data_loop["hour"].unique())
```

```python
load_hour_list=[]

for h in load_data_loop["hour"]:
    if isinstance(h,str) and h.startswith("h"):
        load_hour_list.append(int(h[1:]))
    else:
        load_hour_list.append(None)

load_data_loop["hour"]=load_hour_list
```

```python
weather_hour_list=[]

for h in weather_data_loop["hour"]:
    if isinstance(h,str) and h.startswith("h"):
        weather_hour_list.append(int(h[1:]))
    else:
        weather_hour_list.append(None)

weather_data_loop["hour"]=weather_hour_list
```

```python
load_data_loop["datetime"] = pd.to_datetime(load_data_loop[["year", "month", "day"]]) + pd.to_timedelta(load_data_loop["hour"] - 1, unit="h")
weather_data_loop["datetime"] = pd.to_datetime(weather_data_loop[["year", "month", "day"]]) + pd.to_timedelta(weather_data_loop["hour"] - 1, unit="h")
```

```python
load_data_loop
```

```python
# Combine the two datasets on datetime
df = pd.merge(load_data_loop, weather_data_loop[["datetime", "temperature"]], on="datetime")

# Remove rows with missing values
df = df.dropna()

# Look at the result
df.head()
```

```python
# Add extra time-related columns
df["weekday"] = df["datetime"].dt.day_name()
df["hour"] = df["datetime"].dt.hour
df["month"] = df["datetime"].dt.month
```

```python
#displaying data
print("Data:", df)
```

```python
# %%
import matplotlib.pyplot as plt

# Plot hourly electric load over time
plt.figure(figsize=(15, 5))
plt.plot(load_data_loop["datetime"], load_data_loop["load_MW"], color="blue", alpha=0.7)

plt.title("Hourly Electric Load Over Time", fontsize=14)
plt.xlabel("Date and Time", fontsize=12)
plt.ylabel("Load (MW)", fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

1:- We can see load increases in mid-year(likely summer) and decreases in early months(likely winter)
    This suggest a possible Seasonal Pattern
2:-There are repeating peaks and troughs which suggest yearly trends.
3:- Peak Load reaches ~30000 MW which suggest very high Demand
4: 


```python

# Average load by hour of day
hourly_avg=load_data_loop.groupby("hour")["load_MW"].mean()
plt.figure(figsize=(10, 4))
plt.plot(hourly_avg.index, hourly_avg.values, marker='o',color='green')
plt.title("Average Load by Hour of Day")
plt.xlabel("Hour")
plt.ylabel("Average Load (MW)")
plt.grid(True)
plt.xticks(range(0, 24))
plt.tight_layout()
plt.show()
```

1:-Lowest Load Occurs Early Morning between (2-5 am). This suggests that activity and electricty usage is quiet low.
2:-Load Begin to rise from 6 am, peak around 8-9 am.
3:- Between 10am to 4pm the load is quiet stable.
4:- A significant rise occurs after 5 pm reaching the maximum average load around 7-9 pm.
